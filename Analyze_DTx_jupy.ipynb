```python
##########masking
df['masking'] = df['masking'].replace({'Open label': 'NP', 'NP': 'NP'})
df['masking'] = df['masking'].replace({'open label': 'NP', 'NP': 'NP'})
df['masking'] = df['masking'].apply(lambda x: 'Other' if x != 'NP' else x)
cross_table = pd.crosstab(df['registered'], df['Masking'])
print(cross_table)
##########randomization

df['randomization'] = df['randomization'].replace({'Non-Randomized': 'NP', 'NP': 'NP'})
df['randomization'] = df['randomization'].apply(lambda x: 'Other' if x != 'NP' else x)
cross_table = pd.crosstab(df['registered'], df['randomization'])
print(cross_table)
```
